```python
# Jangan lupa download chromedriver trus tempatin di directory 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd

# Set the path to the ChromeDriver executable
webdriver_path = "path file chromedriver kalian"  # ganti path file chromdriver kalian

# Set up the ChromeDriver service
service = Service(webdriver_path)

# Set up the Chrome options?
options = webdriver.ChromeOptions()

options.add_argument("--incognito") #buat buka browser incognito
options.add_argument('--disable-notifications') #disable notif twitter kalo pertama buka 
options.add_argument("--pageLoadStrategy=none")

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://twitter.com/i/trends")

# Ini ambil datanya dari website di atas
trends = driver.find_elements(By.XPATH, '//div[@dir="ltr"]//span[@class="css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0"]')

# Looping datanya terus masukin ke list
data_trend = []
for trend in trends:
    data_trend.append(trend.text)

# Ambil yang dibutuhin aja     
data_trend = data_trend[5:]

# Split datanya 
def split_list(lst, size):
    return [lst[i:i+size] for i in range(0, len(lst), size)]
my_list = data_trend
sublists = split_list(my_list, 3)

# Tambahin nama kolom biar cakep
columns = ['Location', 'Trending', 'Volume']
data = pd.DataFrame(sublists, columns=columns)
driver.quit()
data = data[:20]

# kalo mau save datanya 
data.to_excel("trend_twitter.xlsx")
data



# kodingan dibawah ini dibuat kalau ada trending topic yang mengadung emoticon hehe
#divider = [x for x in data_trend if 'Trending' in x]

#sublists = [[]]
#for item in data_trend:
#    if any(div in item for div in divider):
#        sublists.append([])
#    sublists[-1].append(item)

#print("Sublists:", sublists)

#coba = sublists[1:20]

#pd.DataFrame(coba)
```
